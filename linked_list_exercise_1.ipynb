```python
class Node:
  def __init__(self, data):
    self.data = data
    self.next = None

class LinkedList:
  def __init__(self):
    self.head = None

  def append(self,data):
    mikUdayOu = Node(data)

    if not self.head:
      self.head = mikUdayOu

    temp = self.head
    while temp.next:
      temp = temp.next

    temp.next = mikUdayOu

  def display(self):
    temp = self.head

    while temp:
      print(temp.data, end="  ->  ")
      temp = temp.next

```
