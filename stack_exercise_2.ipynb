# Stack Using Linked List

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class StackLL:
    def __init__(self):
        self.top = None
    
    def push(self, data):
        new_node = Node(data)
        new_node.next = self.top
        self.top = new_node
    
    def display(self):
        temp = self.top

        while temp:
            print(temp.data, end="  ->  ")
            temp = temp.next
        
        print("None")
```

```python
s = StackLL()

s.push(1)
s.push(2)
s.push(3)

s.display()
```
