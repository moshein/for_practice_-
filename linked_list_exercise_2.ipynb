```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
    
class LinkedList:
    def __init__(self):
        self.head = None
    
    def append(self, data):
        print("=== APPEND START ===")
        print(f"Passed 1: Creating new node -> this is '[{data}]' \n")

        new_node = Node(data)

        if not self.head:
            print("Passed 2: List is empty")
            print(f"    Setting head -> '[{new_node.data}]' node")

            self.head = new_node

            print("=== APPEND END === \n")
            print("==================================================")
            
            return
        
        print("Passed 2: List is NOT empty")
        temp = self.head
        print(f"    Starting traversal from head -> [{temp.data}] node")

        step = 1

        while temp.next:
            print(f"    Step [{step}]: [{temp.data}] node -> moving to [{temp.next.data}] node")
            temp = temp.next
            step += 1
        
        print(f"Passed 3: Reached last node -> [{temp.data}] node")
        print(f"    Linking [{temp.data}] node -> next to [{new_node.data}] node")
        
        temp.next = new_node

        print("=== APPEND END === \n")
        print("==================================================")


```

```python
ll = LinkedList()

ll.append("miku")
ll.append("hatsune")
ll.append("Jesus")
```
