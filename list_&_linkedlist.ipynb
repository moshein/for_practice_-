# Linked List using List

```python
class LinkedList:
    def __init__(self):
        self.list = []
    
    def append(self, data):
        return self.list.append(data)
    
    def remove(self, key):
        return self.list.remove(key)
    
    def sort(self):
        return self.list.sort()
```

```python
listA = LinkedList()

listA.append(1)
listA.append(2)
listA.append(3)

print(listA.list)
```

# Linked List using Linked List itself

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None
    
    def append(self, data):
        new_node = Node(data)

        if not self.head:
            self.head = new_node
            return
        
        temp = self.head

        while temp.next:
            temp = temp.next
        
        temp.next = new_node
    
    def pop(self):
        temp = self.head

        while temp.next.next: ### how come this line is the 2nd-to-last node
            temp = temp.next
        
        temp.next = None
```
