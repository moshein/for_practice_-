```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
```

```python
class StackLL:
    def __init__(self, capacity):
        self.top = None
        self.capacity = capacity
        self.count = 0

    def push(self, data):
        if self.is_full():
            print("Stack is full")
            return

        new_node = Node(data)
        new_node.next = self.top
        self.top = new_node
        self.count += 1

    def pop(self):
        if self.top is None:
            raise IndexError("HI LANGLANG")
        popped_node = self.top
        self.top = self.top.next
        self.count -= 1
        return popped_node.data

    def peek(self):
        if self.top is None:
            return None
        return self.top.data

    def is_empty(self):
        return self.top is None

    def is_full(self):
        if self.count == self.capacity:
            return True
        else:
            return False
```

```python
## Enter capacity

sll = StackLL(3)
```

```python
sll.push(10)
sll.push(20)
sll.push(30)
sll.push(40)
```
