```python
transactions = ["apple", "banana", "apple", "orange", "banana", "apple"]

fin_trans = {}

for i in transactions:
  if i not in fin_trans:
    fin_trans[i] = 1
  else:
    fin_trans[i] += 1

print(fin_trans)
```
