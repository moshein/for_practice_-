## Linked List
image.png </br></br>
A LinkedList is a linear data structure where elements are stored in nodes. </br></br> Each node contains two parts:
* Data
* Next (or pointer) to the next node


```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def insert(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

    def display(self):
        temp = self.head
        while temp:
            print(temp.data, end=" -> ")
            temp = temp.next
        print("None")


ll = LinkedList()
ll.insert(1)
ll.insert(2)
ll.insert(3)
ll.display()
```

## Stack
A Stack is a data structure that follows the LIFO principle — Last In, First Out. </br></br>
image.png </br>


```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


class Stack:
    def __init__(self):
        self.top = None

    def push(self, data):
        new_node = Node(data)
        new_node.next = self.top
        self.top = new_node

    def pop(self):
        if self.top is None:
            return "Stack is empty"

        removed = self.top
        self.top = self.top.next
        return removed.data

    def display(self):
        current = self.top
        while current:
            print(current.data, end=" -> ")
            current = current.next
        print("None")



s = Stack()
s.push(10)
s.push(20)
s.push(30)
s.display()
s.pop()
s.display()
```

## Queue
image.png </br></br>
A Queue follows the FIFO principle — First In, First Out. <br></br>
Think of it like a line:
* The first person in line is the first to leave


```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


class Queue:
    def __init__(self):
        self.front = None
        self.rear = None

    def enqueue(self, data): ## adds to the rear (similar to append)
        new_node = Node(data)

        if self.rear is None: ## if walay sulod ang linkedlist
            self.front = self.rear = new_node
            return

        self.rear.next = new_node
        self.rear = new_node

    def dequeue(self): ## removes from the front (similar when deleting an item)
        if self.front is None:
            return "Queue is empty"

        removed = self.front
        self.front = self.front.next

        if self.front is None:
            self.rear = None

        return removed.data

    def display(self):
        current = self.front
        while current:
            print(current.data, end=" -> ")
            current = current.next
        print("None")



q = Queue()
q.enqueue(1)
q.enqueue(2)
q.enqueue(3)
q.display()
q.dequeue()
q.display()
```
